# Filter Banks of Discrete Wavelet

```python
wavelet = pywt.Wavelet('sym2')
print(wavelet)
```

```python
[dec_low, dec_high, rec_low, rec_high] = wavelet.filter_bank
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, figsize=(12,8))

ax1.stem(dec_low)
ax1.set_title('Decomposition low-pass Filter', fontsize = 15)
ax2.stem(dec_high)
ax2.set_title('Decomposition high-pass Filter', fontsize = 15)
ax3.stem(rec_low)
ax3.set_title('Reconstruction low-pass Filter', fontsize = 15)
ax4.stem(rec_high)
ax4.set_title('Reconstruction High-pass Filter', fontsize = 15)
plt.show()
```

# Decomposition and Reconstruction of Chirp Signal

```python
t = np.linspace(0, 1, 1000)
chirpSignal = np.sin(500 * np.pi * t**2)
```

```python
plt.figure(figsize = (15,4))
plt.plot(chirpSignal, 'm', lw = 3)
plt.xlabel('time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.title("Chirp Signal", fontsize = 15)
plt.show()
```

## Single Level Decomposition

```python
(cA,cD) = pywt.dwt(chirpSignal,'sym2',mode = 'symmetric')
```

```python
plt.figure(figsize = (15,12))
plt.subplot(3,1,1)
plt.plot(chirpSignal, 'm')
plt.title("Chirp Signal", fontsize = 15)
plt.subplot(3,1,2)
plt.plot(cA, 'b')
plt.title("Approximation Coefficients", fontsize = 15)
plt.subplot(3,1,3)
plt.plot(cD, 'g')
plt.title("Detail Coefficients", fontsize = 15)
plt.show()
```

# Signal Reconstruction

```python
reconstrutedSignal = pywt.idwt(cA, cD, 'sym2', mode = 'symmetric')
```

```python
plt.figure(figsize = (15,10))
plt.subplot(2,1,1)
plt.plot(chirpSignal, 'm')
plt.title("Original Signal", fontsize = 15)
plt.subplot(2,1,2)
plt.plot(reconstrutedSignal, 'y')
plt.title("Reconstructed Signal", fontsize = 15)
plt.show()
```

# Multilevel Decomposition

```python
import matplotlib.pyplot as plt 
from matplotlib import style
import numpy as np
import pandas as pd
style.use('dark_background')
import pywt
```

```python
t = np.linspace(0, 1, 1000)
chirpSignal = np.sin(500 * np.pi * t**2)
```

```python
plt.figure(figsize = (15,4))
plt.plot(chirpSignal, 'm', lw = 3)
plt.xlabel('time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.title("Chirp Signal", fontsize = 15)
plt.show()
```

```python
fig, ax = plt.subplots(5, 2, figsize = (10,7))
for ii in range(5):
    (chirpSignal, cD) = pywt.dwt(chirpSignal, 'sym5')
    ax[ii, 0].plot(chirpSignal, 'b')
    ax[ii, 1].plot(cD, 'g')
    ax[ii, 0].set_ylabel("Level {}".format(ii + 1), fontsize=14, rotation=90)
    ax[ii, 0].set_yticklabels([])
    if ii == 0:
        ax[ii, 0].set_title("Approximation coefficients", fontsize=14)
        ax[ii, 1].set_title("Detail coefficients", fontsize=14)
    ax[ii, 1].set_yticklabels([])
plt.tight_layout()
plt.show()
```

# Time Frequency Analysis With Continuous Wavelet Transform

# Importing Libraries

```python
import matplotlib.pyplot as plt 
from matplotlib import style
import numpy as np
import pandas as pd
style.use('dark_background')
import pywt
```

# Reading a SIgnal

```python
spike = pd.read_csv('Spike.csv')
```

```python
spike.shape
```

```python
signal = spike[0:400]
```

```python
plt.figure(figsize = (12,6))
plt.plot(signal, 'm', lw = 3)
```

# Generate Morlet Wavelet

```python
srate = 1024 
t = np.arange(0.,5.,1/srate) # time vector in seconds

# centered time vector

t = t - np.mean(t)
```

```python
# parameter setting
f = 4 #Hz
csw  = np.cos(2*np.pi*f*t) 

fwhm = .5 #  seconds
gaussian = np.exp( -(4*np.log(2)*t**2) / fwhm**2 ) 

# Morlet wavelet
MorletWavelet = csw * gaussian
```

```python
plt.figure(figsize = (12,6))
plt.plot(t,MorletWavelet,'m', lw =3)
plt.xlabel('Time')
plt.title('Morlet wavelet', fontsize = 20)
plt.show()
```

# Computing Continuous Wavelet Transform

```python
coef,_ = pywt.cwt(signal, scales = np.arange(1,10), wavelet = 'morl')
```

```python
plt.figure(figsize=(20, 7))
plt.imshow(abs(coef), extent=[0, 400, 1, 10], interpolation='bilinear', aspect='auto')
plt.yticks(np.arange(1, 11, 1))
plt.xticks(np.arange(0, 401, 10))
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Frequency', fontsize = 15)
plt.title('Time Frequency Analysis', fontsize = 20)
plt.show()
```
