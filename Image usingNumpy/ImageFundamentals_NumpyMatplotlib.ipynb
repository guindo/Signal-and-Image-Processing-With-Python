```python
import numpy as np
import matplotlib.pyplot as plt
```

# Reading an RGB Image

```python
img1 = plt.imread('Vegetable.jpg')
```

```python
plt.imshow(img1)      # R,G,B format
plt.title('Vegetable')
plt.show()
```

```python
print(type(img1))
print(" ")
print(img1.shape)
print(" ")
print(img1.ndim) # Number of dimension. The third value shows the channels
print(" ")
print(img1.size) # 225 x 225 x 3 = 151875
print(" ")
print(img1.dtype)
```

# Saving an Image

```python
plt.imsave('MyImage.png', img1)
```

# Reading and displaying Gray scale Image

```python
img2 = plt.imread('Flower.jpg') # When matplotlib read grayscaleimage, it looks for color componenets. Thus, it displays the
                                # gray scale image with some random combination which is not the true representation of image
plt.imshow(img2)
```

```python
plt.imshow(img2, cmap='gray')
plt.show()
```

```python
print(type(img2))
print(" ")
print(img2.shape)
print(" ")
print(img2.ndim)
print(" ")
print(img2.size)
print(" ")
print(img2.dtype)
```

```python
img3 = img1.astype(np.float32)/255
```

```python
print(type(img3))
print(" ")
print(img3.shape)
print(" ")
print(img3.ndim)
print(" ")
print(img3.size)
print(" ")
print(img3.dtype)
```

# Observing the red, green and blue components of color (RGB) image

```python
plt.imshow(img1)     
plt.show()
```

```python
print(img1.shape)
```

```python
plt.imshow(img1[:,:,0], cmap = 'gray') # see the red
```

```python
plt.imshow(img1[:,:,1], cmap = 'gray') # see the green
```

```python
plt.imshow(img1[:,:,2], cmap = 'gray') # see the blue
```

# Finding Pixel Values of red, green and blue componenets at [100,100]

```python
print(img1[100, 100, 0]) # red value
print(img1[100, 100, 1]) # green value
print(img1[100, 100, 2]) # blue value
```

```python
print(img1[100, 100, :]) # see all values (R,G,B)
```
