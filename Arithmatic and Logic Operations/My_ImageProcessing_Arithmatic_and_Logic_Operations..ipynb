# Addition

```python
img1 = cv2.imread('lena.tif', 0)
```

```python
print(type(img1))
print(" ")
print(img1.dtype)
print(" ")
print(img1.shape)
```

```python
plt.figure(figsize=(12,6))
plt.imshow(img1, cmap = 'gray')
plt.show()
```

```python
print(img1)
```

```python
scalar = np.ones(img1.shape, dtype = "uint8") * 100
```

```python
print(scalar)
```

# Addition Using CV2

```python
addimg1 = cv2.add(img1, scalar)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img1, cmap = 'gray')
plt.title("Original Image", fontsize = 15)
plt.subplot(1,2,2)
plt.imshow(addimg1 , cmap = 'gray')
plt.title("Image After adding an Scalar", fontsize = 15)
plt.show()
```

```python
print(addimg1)
```

# Addition Using Numpy

```python
addimg2 = img1 + scalar
plt.figure(figsize=(12,6))
plt.imshow(addimg2, cmap = 'gray')
plt.title("Numpy produce results after clipping", fontsize = 15)
plt.show()
```

```python
print(addimg2)
```

# Observing Difference between CV2 Addition and Numpy Addition

```python
img1[40:50, 100:120]     # Original Image 
```

```python
addimg1[40:50, 100:120]    # Image Addition by cv2
```

```python
addimg2[40:50, 100:120]  # Image Addition by NumPy
```

# Subtraction

```python
sub_img1 = cv2.subtract(img1, scalar)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img1, cmap = 'gray')
plt.title("Original Image", fontsize = 15)
plt.subplot(1,2,2)
plt.imshow(sub_img1 , cmap = 'gray')
plt.title("Image After Subtraction", fontsize = 15)
plt.show()
```

```python
print(sub_img1)
```

```python
sub_img2 = img1 - scalar
plt.figure(figsize=(12,6))
plt.imshow(sub_img2, cmap = 'gray')
plt.title("Numpy produce results after clipping", fontsize = 15)
plt.show()
```

```python
print(sub_img2)
```

# Logical Operations

```python
X1 = cv2.imread('X1.png', 0)
```

```python
print(type(X1))
print(" ")
print(X1.dtype)
print(" ")
print(X1.shape)
```

```python
X1 =cv2.resize(X1,(250,210))
```

```python
plt.figure(figsize=(8,4))
plt.imshow(X1, cmap = 'gray')
plt.title('$X_1$', fontsize = 15)
plt.show()
```

```python
X2 = cv2.imread('X2.png', 0)
```

```python
X2 =cv2.resize(X2,(250,210))
```

```python
plt.figure(figsize=(8,4))
plt.imshow(X2, cmap = 'gray')
plt.title('$X_2$', fontsize = 15)
plt.show()
```

## Logical and

```python
and_Image = cv2.bitwise_and(X1,X2)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(X1, cmap = 'gray')
plt.title("$X_1$", fontsize = 15)
plt.subplot(1,3,2)
plt.imshow(X2 , cmap = 'gray')
plt.title("$X_2$", fontsize = 15)
plt.subplot(1,3,3)
plt.imshow(and_Image , cmap = 'gray')
plt.title("$X_1 and x_2$", fontsize = 15)
plt.show()
```

# Logical or

```python
or_Image = cv2.bitwise_or(X1,X2)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(X1, cmap = 'gray')
plt.title("$X_1$", fontsize = 15)
plt.subplot(1,3,2)
plt.imshow(X2 , cmap = 'gray')
plt.title("$X_2$", fontsize = 15)
plt.subplot(1,3,3)
plt.imshow(or_Image , cmap = 'gray')
plt.title("$X_1 or x_2$", fontsize = 15)
plt.show()
```

# Logical xor

```python
xor_Image = cv2.bitwise_xor(X1,X2)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(X1, cmap = 'gray')
plt.title("$X_1$", fontsize = 15)
plt.subplot(1,3,2)
plt.imshow(X2 , cmap = 'gray')
plt.title("$X_2$", fontsize = 15)
plt.subplot(1,3,3)
plt.imshow(xor_Image , cmap = 'gray')
plt.title("$X_1 xor x_2$", fontsize = 15)
plt.show()
```

# Logical not

```python
not_X1 = cv2.bitwise_not(X1) # Also a negative image
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(X1, cmap = 'gray')
plt.title("$X_1$", fontsize = 15)
plt.subplot(1,2,2)
plt.imshow(not_X1 , cmap = 'gray')
plt.title("$not X_1$", fontsize = 15)
plt.show()
```

```python
not_X2 = cv2.bitwise_not(X2)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(X2, cmap = 'gray')
plt.title("$X_2$", fontsize = 15)
plt.subplot(1,2,2)
plt.imshow(not_X2 , cmap = 'gray')
plt.title("$not X_2$", fontsize = 15)
plt.show()
```
